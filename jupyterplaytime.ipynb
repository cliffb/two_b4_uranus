```python
import pandas as pd
import io
import requests
import matplotlib.pyplot as plt

url="https://app.periscopedata.com/api/shippingeasy/chart/csv/bdea7643-7d7a-54a1-dba8-1b9e69789101"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
c['day'] = pd.to_datetime(c['day'],infer_datetime_format=True) #converts dates to useable axis values

#c.plot(kind='line',x='day',y='adoptions')
#c.plot(kind='line',x='day',y='signups')
#plt.show()


plt.plot(c.day, c.adoptions, '-.', color="#333333", label="adoptions")
plt.plot(c.day, c.signups, '--', color="#999999", label="signups")
plt.plot(c.day, c.activations, '-', color="#aaaaaa", label="activations")
plt.legend(loc="upper left")

plt.title("Std Paid Signups", fontsize=16, fontweight='bold')
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.show()
```
